```python
from keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
def plot_samples(data, title, wide=3,tall=3):
    """
    helper to plot and save mnist samples
    input - seed number used for both numpy and tensorflow
    """
    subplot_no = 1
    for i in data:
        plt.subplots_adjust(hspace=0.8, wspace=0.8)
        plt.subplot(3,3,subplot_no)
        subplot_no = subplot_no +1
        plt.suptitle(title)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(i, cmap='Greys')    
        plt.savefig('output/' + title + '.png', bbox_inches='tight')
```

```python
sample_indices = [1, 2, 4, 8, 16 , 32, 65, 1329,4000]
plot_samples(X_train[sample_indices], 'mnist samples')
```

```python
# 3. Reshape to 4D vector for Keras compability

X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2], 1).astype('float32')
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2], 1).astype('float32')
print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
# 4. Cast data to float32 and 
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# 4. normalize inputs from 0-255 to 0-1
X_train/=255
X_test/=255

```

```python
from keras.utils import to_categorical

# 4. convert class labels to categorical data/one-hot encoding

y_test = to_categorical(y_test)
y_train = to_categorical(y_train)
```

```python
# 5. sequential neural network
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(400, input_dim=28*28, activation='sigmoid'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

The softmax activation has the property that all the activations sum to 1, and lie between 0 and 1. It gives a probability distribution rather than a definite answer for a data point, which represents better the uncertainty of the classification. The softmax activation is appropriate here because the classes are mutually exclusive, a digit can belong to only one class.

The first hidden layer has 400 nodes, each connected to every pixel in the input data. There is also a bias node, making the total 401 x 28 x 28 = 314000.

The output layer has 10 nodes connected to each of the 401 outputs from the first node, for a total of 4010 weights.

```python
from keras import optimizers
model.compile(loss='categorical_crossentropy', optimizer=optimizers.SGD(), metrics=['accuracy'])
```

```python
X_train.reshape(-1,28*28)
```

```python
results = model.fit(X_train.reshape(-1,28*28), y_train, epochs=10, batch_size=64, validation_split=1/12)
```

```python
def training_eval(results):
    """
    generate plots of training and validation loss for network training
    
    return -    a keras model that can be used as an input
                to the predict and train functions specified below
    	
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    sns.set()    
    # summarize history for accuracy
    plt.clf()
    plt.plot(results.history['acc'])
    plt.plot(results.history['val_acc'])
    plt.title('simple classifier training accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['training', 'validation'], loc='upper left')
    plt.savefig('output/simple_training_accuracy.png', bbox_inches='tight')
    plt.savefig('output/simple_training_accuracy.pdf', bbox_inches='tight')
    
    # summarize history for loss
    plt.clf()
    plt.plot(results.history['loss'])
    plt.plot(results.history['val_loss'])
    plt.title('simple classifier training loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['training', 'validation'], loc='upper left')
    plt.savefig('output/simple_training_loss.png', bbox_inches='tight')
    plt.savefig('output/simple_training_loss.pdf', bbox_inches='tight')
```

```python
training_eval(results)
```

```python
y_fit = model.predict(X_test.reshape(-1,28*28), batch_size=128)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def performance_eval(title, y_fit, y_target):
    """
    reports on performance of a given data set vs a given target
    input - title, string to be used for labeling plots and text files
            y_fit, output from model
            y_target, target classes of the data
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    report = classification_report(y_target, y_fit)
    with open('output/' +title + '_classification.txt', 'w') as file:
        file.write(report)
    print(report)
    fig, ax = plt.subplots(figsize=(10,10))
    mat = confusion_matrix(y_target, y_fit)
    acc = accuracy_score(y_target,y_fit)
    accuracy_report = str('\n\n{} accuracy: {:.5f}'.format(title, acc))
    with open('output/' +title + '_classification.txt', 'a') as file:
        file.write(accuracy_report)
    print(accuracy_report)
    sns.set()
    sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, cmap='BuGn')
    plt.xlabel('true label')
    plt.ylabel('predicted label')
    plt.savefig('output/' + title + '_confusion.png', bbox_inches='tight')
    plt.savefig('output/' +title + '_confusion.pdf' ,bbox_inches='tight')
```

```python
import numpy as np

performance_eval('simple', y_fit.argmax(axis=1), y_test.argmax(axis=1))
```

```python
def save_summary(model, title):
    with open('models/'+title+'_summary.txt', 'w') as f:
        model.summary(print_fn=lambda x: f.write(x + '\n'))
    print(model.summary())
```
