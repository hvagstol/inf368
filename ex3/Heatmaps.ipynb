```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(14,10)})

for i in range(1,21):
    df = pd.read_csv('clusters.'+str(i)+'.log', delim_whitespace=True, header=None,index_col=0)
    df.columns= df.index
    fig = sns.heatmap(df, fmt="d", annot=True,cmap='BuGn', linewidths=1, linecolor='white')
    fig.set(xlabel='Predicted Class', ylabel='True Class', title='Iteration '+str(i) +'\nClosest Centroid - Confusion Matrix')
    plt.savefig('val_confusion_'+str(i)+'.png')
    plt.clf()
    
```
